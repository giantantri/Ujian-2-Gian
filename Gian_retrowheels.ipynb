```python
kursor=mydb.cursor()
```

```python
def buat_sql (query):
    kursor.execute(query)
    hasil=kursor.fetchall()
    data=pd.DataFrame(data=hasil,columns=kursor.column_names)
    return data
```

```python
buat_sql ("select distinct count(customerNumber) as Customers, count(city) as Cities, count(country) as Countries from customers")
```

```python
buat_sql("select distinct count(emp.employeeNumber) as Employee , count(office.officeCode) as Offices, count(office.country) as Country , count(product.productCode) as Products , count(product.quantityInStock) as StockProducts , count(product.productVendor) as Vendors from employees as emp left join offices as office on emp.officeCode = office.officeCode left join customers as cust on cust.city = office.city left join orders as orders on orders.customerNumber = cust.customerNumber left join orderdetails as od on od.orderNumber = orders.orderNumber left join products as product on product.productCode = od.productCode;")
```

```python
buat_sql("select product.productLine as productLine, min(product.buyPrice) as minPrice , max(product.buyPrice) as maxPrice from products as product group by product.productLine;")
```

```python
buat_sql("select customerName as customerName, city as city, country as country, sum(creditLimit) as total from customers group by customerName order by creditLimit desc limit 10;")
```
